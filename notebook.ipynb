## Data Wrangling

### Gathering Data

```python
day_df = pd.read_csv("./data/day.csv")
hour_df = pd.read_csv("./data/hour.csv")

#melihat dataframe harian
day_df.describe(include="all")
```

### Assessing Data

```python
# Melihat row dimana total penyewaan hanya 22 buah, untuk mengecek peristiwa di hari tersebut
day_df[day_df.cnt == 22]
```

```python
# Menentukan nilai tersebut merupakan outlier atau tidak

#Menggunakan IQR
q25 = day_df.cnt.quantile(0.25)
q75 = day_df.cnt.quantile(0.75)

iqr = q75 - q25
cut_off = iqr * 1.5

minimum = q25 - cut_off
maximum = q75 + cut_off

print("Nilai minimum = ", minimum)
print("Nilai maximum = ", maximum)

outliers = [x for x in day_df.cnt.tolist() if x < minimum or x > maximum]

print("Jumlah outlier : ", outliers)
#List kosong

# Menggunakan Boxplot
day_df.boxplot(column="cnt")
plt.show()
# Whisker boxplot berhenti pada nilai maksimum dan minimum cnt, bukan +/- cutoff
```

```python
# Melihat row dimana kolum kelembaban 0%, untuk mengecek peristiwa di hari tersebut.
day_df[day_df.hum == 0]
# Tidak ada peristiwa yang berhubungan dengan kelembaban 0%
```

```python
#Mengecek kelembaban hari sebelum dan sesudah dimana hum = 0%
print(hour_df[(hour_df.dteday >= "2011-03-09") & (hour_df.dteday <= "2011-03-11")].to_string())
```

```python
#Melihat jika ada kolum yang kosong
day_df.info()

#Melihat juka ada duplikat
print("Jumlah duplikat : ", day_df.duplicated().sum())
```

**Insight:**
- Jumlah "cnt" 22 bukan outlier.

- Jumlah data lengkap, tidak ada yang kosong/null
- Data tidak ada yang duplikat

### Cleaning Data

```python
#mereplace nilai 0 dengan nan
hour_df.hum = hour_df.hum.replace(to_replace=0,value=np.nan)

#Mengisi nan dengan mean bulan itu
mean_bulan = hour_df.hum[(hour_df.dteday <= "2011-03-31") & (hour_df.dteday >= "2011-03-01")].mean()
hour_df.hum = hour_df.hum.fillna(value=mean_bulan)
print(hour_df[(hour_df.dteday >= "2011-03-09") & (hour_df.dteday <= "2011-03-11")].to_string())
```

```python
#Baca insight terlebih dahulu
print(day_df[(day_df.dteday <= "2012-11-30") & (day_df.dteday >= "2012-09-01")].to_string())
```

**Insight:**
- hum yang bernilai 0, diisi dengan bulan nilai mean bulan itu.
- row cnt yang memiliki nilai 22 sya tidak bersihkan karena efeknya yang berkesinambungan. Adanya efek "Regression toward the mean", nilai ekstrem (22) cenderung membuat nilai kembali ke rata-rata. Sebelum adanya musibah Hurrican Sandy cnt bisa mencapai 8 ribuan sedangkan setelah musibah tersebut hanya bisa mencapai 6 ribuan. Kita bisa lihat rata-rata harian cnt pada bagian gathering data adalah 4504.348837, nilai 6 ribu tersebut mendekati mean.

## Exploratory Data Analysis (EDA)

### Explore ...

```python
#Mengecek korelasi kolum season, holiday, hr, weekday, workingday, weathersit, temp, atemp, hum, windspeed dengan cnt
#Menggunakan perjam karena memiliki lebih banyak data
hour_df_copy = hour_df.copy()
hour_df_copy = hour_df_copy.drop(columns=["instant", "dteday", "yr", "mnth", "casual", "registered"])
hour_df_copy.corr()
```

```python
#membuat dataframe bulanan
#mengganti tiped data dteday dari string menjadi datetime
day_df.dteday = pd.to_datetime(day_df.dteday)

#membuat dataframe baru dengan jumlah cnt, me-mean-kan temp dan atemp dalam satu bulan
monthly_df = day_df.resample(rule='ME', on='dteday').agg({
    "temp": "mean",
    "atemp": "mean",
    "cnt": "sum"
})
monthly_df.index = monthly_df.index.strftime('%Y-%m')
monthly_df = monthly_df.reset_index()
print(monthly_df.to_string())
```

**Insight:**
- season ~ cnt = lemah positif
- hr ~ cnt = sedang positif
- holiday ~ cnt = lemah negatif
- weekday ~ cnt = lemah positif
- workingday ~ cnt = lemah positif
- weathersit ~ cnt = lemah negatif
- temp ~ cnt = sedang positif
- atemp ~ cnt = sedang positif
- hum ~ cnt = sedang negatif
- windspeed ~ cnt = lemah positif

## Visualization & Explanatory Analysis

### Pertanyaan 1: Pada kondisi apa peminjaman sepeda paling banyak dan paling sedikit ?

```python
sns.scatterplot(data = hour_df, x = "hr", y = "cnt")
plt.show()

sns.scatterplot(data = hour_df, x = "temp", y = "cnt")
plt.show()

sns.scatterplot(data = hour_df, x = "atemp", y = "cnt")
plt.show()

sns.scatterplot(data = hour_df, x = "hum", y = "cnt")
plt.show()
```

### Pertanyaan 2: Bagaimana perkembangan sistem bike sharing dalam kurun dua tahun ?

```python
plt.figure(figsize=(10, 5)) 
plt.plot(monthly_df["dteday"], monthly_df["cnt"], marker='o', linewidth=2, color="#72BCD4") 
plt.title("Perkembangan Sistem Bike Sharing Dalam Kurun Dua Tahun", loc="center", fontsize=18) 
plt.xticks(fontsize=8, rotation=45) 
plt.yticks(fontsize=8) 
plt.show()
```

**Insight:**
- Penyewaan terbanyak bisa terjadi dengan kondisi berikut:
    - Jam 8, 17, dan 18
    - temp sekitar 0.6 sampai 0.7
    - atemp sekitar 0.6 sampai 0.7
    - hum sekitar 0.3 sampai 0.5
- Penyewaan paling sedikit bisa terjadi dengan kondisi berikut:
    - Jam 1 sampai 4
    - temp sekitar 0.0 sampai 0.1
    - atemp sekitar 0.0 sampai 0.1
    - hum sekitar 0.1 sampai 0.2
      
- atemp memiliki penurunan tajam sekitar 0.3 sampai 0.4 dan 0.5 sampai 0.6. Ini mungkin karena hasil pembulatan.
- Sistem bike sharing memiliki naik diawal tahun dan menurun di akhir tahun. Ini terjadi karena musim dingin yang relatif memiliki suhu rendah, kita bisa lihat di grafik temp, dimana suhu rendah mempunyai cnt yang sedikit.
- Jumlah max cnt naik dalam kurun dua tahun.
